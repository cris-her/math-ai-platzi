## Funciones

Las funciones muestran la relación entre dos conjuntos de datos. Por ejemplo, la relación entre el número de horas trabajadas con el sueldo ganado, la sensación de saciedad apartir de la cantidad de comida ingerida. En conclusión una función describe como se relaciona los datos de un conjunto con respecto a otro e indica como obtener uno a partir del otro.

$$ sueldo = horas * 100  \Rightarrow f(h) = s(h) = h * 100 $$

La notación utilizada más arriba nos muestra el significado de la función, como esta definido un sueldo de una persona, 
$sueldo = horas * 100$, luego vemos como se define $s(h)$ donde se reprsenta al **_sueldo_** como $s$ y las **_horas trabajadas_** como $h$, luego la notación $f(h)$ nos indica que la función depende de $h$, a su vez es igual $s(h)$.

### Tipos de variables

En las funciones se tienen al menos dos variables, una será **dependiente** pues su valor depende del valor de la variable **independiente** y la función que define la relación entre estas ambas. Por otro lado, la variable independiente puedes tomar valores eligidos a discresión, puesto que lo que queremos averiguar al valor que toma esta al realizar la operación(es) definidas por la función.  
En nuestro ejemplo anterior las **horas** son la variable independiente y el **sueldo** la dependiente.

### Dominio y Codominio 

El **codominio** (o imagen) lo que busca es expresar el rango de valores que puede tomar nuestra función después de evaluda, es decir, los valores de la variable dependiente. Por otro lado el **dominio** expresa los valores que puede tomar la variable independiente, es decir, el conjunto de datos de partida.  
En nuestro ejemplo tenemos que nuestro _dominio_ sería $h \in \left\{ 0, +\infty \right)$ y por otro lado tendíamos que el _codominio_ sería $h \in \left\{ 0, +\infty \right)$

### Gráficas

Estan buscan mostrar el comportamiento de la función a lo largo de una region de su dominio, de esta manera se facilita el entedimiento de la misma.  
Pensemos por un momento en nuestro ejemplo, imaginemos que queremos saber cuál sería nuestro sueldo si trabajamos 24 horas por semana o 48 por semana, podemos ver el cuanto aumentará nuestro sueldo entre esos valores.

```python
import numpy as np
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
horas = np.arange(start=24, stop=49) #No inclusivo limite superior.
sueldo = 100 * horas
print(f' - Rango de horas (dominio): {horas}')
print(f' - Valores de sueldo obtenidos (codominios): {sueldo}')
```

```python
plt.figure()
plt.title('s(h) = h * 100')
plt.xlabel('Horas (h)')
plt.ylabel('Sueldo (s)')
plt.plot(horas, sueldo,  linewidth=2)

#Se añaden los marcadores rojos
h_marker = np.arange(24, 49, step=4)
markers = 100 * h_marker
plt.plot(h_marker, markers, color='red', marker='o', linestyle='None')
for a,b in zip(h_marker, markers): 
    plt.text(a+1, b, f's({a})= {b}')
```

### Derivada 

El concepto intuido sobre la derivada nos indica que este mide **el cambio en la pendiente de la función lo que es lo mismo**, sería como ver con qué **_velocidad_** pasamos de un valor a otro valor en el **codominio**. Entonces regresando al ejemplo anterior, ¿cuál sería la _velocidad_ (taza de cambio)  con la que obtenemos un sueldo distinto?. Vamos a verlo, sin preocuparnos directamente como si cálculo al derivada de la función (esa información esta disponible en tablas por ejemplo [aquí](http://matematicas.relatividad.org/derivadas.gif)). Recordemos nuestra función y calculemos su derivada, que esta denata como $s'(h)$ :  

$$ s(h) = h * 100  \\ 
s'(h) = 100$$

```python
from sympy import Symbol
from sympy import diff

h = Symbol("h")
s = h * 100
s1 = diff(s, h) #Calcaula la derivada con respecto a 'h'
print(f'Nuestra derivada es: {s1}')
```

Haciendo uso de la librería de `Sympy` que nos permite hacer cálculo simbólico, obtuvimos el mismo resultado. Por otro lado que signifca este 100, para entender ese valor constante podemos ver gráfica de la relación de nuestra derivada para los valores que toma nuestra variable $h$, porque como muestra la notación y como se le indico a `Sympy` derivamos con respecto la varible $h$. 

```python
cambios = np.repeat(s1, len(horas))

plt.figure()
plt.title("s'(h) = 100")
plt.xlabel('Horas (h)')
plt.ylabel("Cambio del sueldo (s')")
plt.plot(horas, cambios,  linewidth=2)

#Se añaden los marcadores rojos
h_marker = np.arange(24, 49, step=4)
markers = np.repeat(s1, len(h_marker))
plt.plot(h_marker, markers, color='red', marker='o', linestyle='None')
switch = 0.8
for a,b in zip(h_marker, markers): 
    plt.text(a-2, b+(switch) , f"s'({a})= {b}")
    switch *= -1
```

Luego esto nos indica que siempre la taza de cambio entre dos valores de sueldo estará dada por el valor que toma $s'(h)$, en este caso $100$. Por ejemplo, cuando $h = 24$ tenemos que $s(24) = 2400$, luego si calculamos el valor de $h = 25$ obtendríamos que $s(25) = 2500$, luego si queremos ver de cuanto fue el cambio entre los valores del sueldo tendríamos qué $s(25) - s(24) = 2500 - 2400 = 100$.
